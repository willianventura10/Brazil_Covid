```python
import pandas as pd
```

```python
#Dados obtidos no Kaggle https://www.kaggle.com/unanimad/corona-virus-brazil
df = pd.read_csv("brazil_covid19_cities.csv")
df = df.drop(columns=['code'])
```

```python
df.head(-5)
```

```python
df = df.groupby(by=["name"]).max()
```

```python
df = df.assign(Population=df.index) 
```

```python
df.head(-5)
```

```python
#Dados obtidos no site do IBGE https://ftp.ibge.gov.br/Estimativas_de_Populacao/Estimativas_2020/estimativa_dou_2020.xls
df2 = pd.read_excel("population_2020.xls")
```

```python
df2.head()
```

```python
lista = list(df2['CITY'])
```

```python
def localiza(nome):
    if nome in lista:
        x = df2.loc[df2['CITY']==nome]
        x = x.iloc[0,1]
        if type(x) == int:
            return (x)
        else:
            return (int(x))
```

```python
df["Population"]=df["Population"].map(localiza)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df2 = df.assign(cases_pc=(df['cases']/df['Population'])) 
df2 = df2.assign(deaths_pc=(df['deaths']/df['Population']))
```

```python
df2.describe()
```

```python
df2.query('name=="São Paulo"')
```

```python
df2['cases'] = df2['cases'].astype(int)
```

```python
df2['Population'] = df2['Population'].astype(int)
```
